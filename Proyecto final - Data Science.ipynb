#### Ingeniería de las caracteristicas


Para comprender mejor cómo las diferencias de género y la presencia de niños influyen en las travesías migratorias, hemos decidido enriquecer nuestro análisis añadiendo nuevas variables que capten estas realidades de forma más detallada. Hemos introducido en nuestro conjunto de datos las siguientes categorías:

"Proportion of Females" y "Proportion of Males": Estas variables nos ayudan a entender el balance de género en cada grupo observado, asignándoles un valor de cero cuando no se registren adultos, para mantener la claridad en nuestros cálculos.

"Children Present": Con un simple sí o no (1 o 0), esta variable indica la presencia de niños en el grupo, lo que nos permite captar de manera directa su participación en estos movimientos.

"Group Size": Este número representa la cantidad total de personas en cada situación, sumando mujeres, hombres, y niños, dándonos una idea clara del tamaño de cada grupo migratorio.

Gracias a estas adiciones, nuestro conjunto de datos ahora tiene una capacidad mejorada para revelar historias importantes sobre la supervivencia y las experiencias de los migrantes, ofreciendo nuevos ángulos desde los cuales podemos explorar los datos y afinar la precisión de nuestros análisis y modelos predictivos.

```python
# Crear nuevas características

# Proporción de Mujeres y Proporción de Hombres
df_limpio['Proportion of Females'] = df_limpio['Number of Females'] / (df_limpio['Number of Females'] + df_limpio['Number of Males'])
df_limpio['Proportion of Males'] = df_limpio['Number of Males'] / (df_limpio['Number of Females'] + df_limpio['Number of Males'])

# Indicador de Presencia de Niños (1 si hay niños, 0 si no)
df_limpio['Children Present'] = (df_limpio['Number of Children'] > 0).astype(int)

# Tamaño del Grupo
df_limpio['Group Size'] = df_limpio['Number of Females'] + df_limpio['Number of Males'] + df_limpio['Number of Children']

# Reemplazar posibles NaN generados por divisiones 0/0 con 0
df_limpio.fillna(0, inplace=True)

# Mostrar las primeras filas para revisar las nuevas características
df_limpio[['Proportion of Females', 'Proportion of Males', 'Children Present', 'Group Size']].head()
```

```python
# Definimos los bins y las etiquetas para los rangos sin la parte 'supervivientes'
bins = [-1, 0, 10, 50, 100, float('inf')]
labels = ['0', '1-10', '11-50', '51-100', '> 100'] 

# Categorizamos 'Number of Survivors' usando pd.cut con las nuevas etiquetas
df_limpio['Survival_Range'] = pd.cut(df_limpio['Number of Survivors'], bins=bins, labels=labels)

# Verificamos los cambios
print(df_limpio['Survival_Range'].unique())
```

Se ha tomado la decisión de organizar la información sobre cuántas personas sobreviven a sus viajes migratorios en grupos claros y definidos, usando una herramienta llamada pd.cut. Este paso es como ordenar un conjunto de notas dispersas en secciones bien marcadas, lo que nos ayuda a entender y explicar mejor cómo varían las posibilidades de supervivencia durante estos difíciles viajes. Al convertir números continuos en categorías más simples, hacemos que sea más fácil para todos - desde quienes analizan los datos hasta quienes toman decisiones basadas en ellos - ver las historias detrás de los números. Este enfoque no solo nos permite hacer predicciones más precisas sobre futuros patrones de supervivencia sino que también asegura que los insights que descubrimos sean útiles y accesibles para quienes buscan mejorar la vida de los migrantes.

En nuestra exploración, enfrentamos una decisión crucial sobre cómo medir el éxito de los viajes migratorios: ¿deberíamos centrarnos en el número exacto de personas que logran sobrevivir o en categorías más generales de supervivencia? Esta elección depende de lo que esperamos descubrir. Si nuestro interés yace en comprender las probabilidades generales de supervivencia basadas en distintas circunstancias, entonces agrupar a los supervivientes en rangos, como lo hemos hecho, se alinea con nuestros objetivos.

```python
import plotly.express as px

# Calcular un estimado de peligrosidad para mujeres multiplicando la proporción de mujeres por el total de muertos y desaparecidos
df_limpio['Estimated Danger for Women'] = df_limpio['Proportion of Females'] * df_limpio['Total Number of Dead and Missing']

# Agrupar por ruta migratoria y sumar el estimado de peligrosidad para mujeres
danger_routes_women = df_limpio.groupby('Combined_Route_Grouping')['Estimated Danger for Women'].sum().reset_index()

# Ordenar las rutas en orden descendente por peligrosidad estimada y seleccionar el top 10
top_danger_routes_women = danger_routes_women.sort_values(by='Estimated Danger for Women', ascending=False).head(10)

# Crear el gráfico de barras con el top 10 de rutas más peligrosas para mujeres
fig4 = px.bar(top_danger_routes_women, 
             x='Estimated Danger for Women', 
             y='Combined_Route_Grouping',
             labels={'Combined_Route_Grouping': 'Migratory Route', 'Estimated Danger for Women': 'Estimated Danger for Women'},
             title="Las 10 rutas migratorias más peligrosas para las mujeres según su proporción",
             orientation='h')

# Mejorar la disposición del gráfico
fig4.update_layout(yaxis={'categoryorder': 'total ascending'})

# Guardar el cuarto gráfico en su propio archivo HTML
file_path_fig4 = "C:/Users/acbon/Desktop/Data Science/TrabajoFinal-DataScience/Rutas_Peligrosas_Mujeres.html"
pio.write_html(fig4, file=file_path_fig4, include_plotlyjs='cdn')
 
fig4.show() 
```

Valores más altos indican un mayor peligro para las mujeres, implicando que en esas rutas se han registrado eventos con alta participación femenina y un elevado número de víctimas.

Valores más bajos sugieren que, aunque haya habido incidentes mortales o desapariciones, la participación de mujeres y/o el número total de víctimas fue menor, resultando en un menor "peligro estimado".

 Por otro lado, RobustScaler se utiliza para las variables numéricas continuas, mitigando el impacto de los valores atípicos y asegurando que los modelos desempeñen de manera óptima. Esta cuidadosa selección de técnicas de codificación y escalado está diseñada para reflejar la complejidad del fenómeno migratorio y maximizar el rendimiento de los modelos avanzados, contribuyendo a resultados precisos y confiables para una mejor comprensión y formulación de políticas públicas efectivas.






```python

# Definir las columnas para cada tipo de transformación
columnas_categoricas_one_hot = ['Reported Month', 'Region of Incident', 'Cause of Death']
columnas_categoricas_frecuencia = ['Location of death', 'Combined_Region_Country', 'Combined_Route_Grouping']
columnas_numericas = ['Incident year', 'Number of Dead', 'Total Number of Dead and Missing', 
                      'Number of Females', 'Number of Males', 'Number of Children', 
                      'Latitude', 'Longitude', 'Proportion of Females', 'Proportion of Males', 
                      'Children Present', 'Group Size', 'Estimated Danger for Women']

# Inicializar el LabelEncoder
encoder = LabelEncoder()

# Ajustar y transformar 'Survival_Range' para convertir las etiquetas de cadena a numéricas
y_encoded = encoder.fit_transform(df_limpio['Survival_Range'])

# Guardar el mapeo de categorías a números
category_mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))

# Mostrar el mapeo
print("Mapeo de categorías a etiquetas numéricas:", category_mapping)

# Codificación one-hot
onehot = OneHotEncoder(sparse=False, handle_unknown='ignore')
data_codificada_one_hot = onehot.fit_transform(df_limpio[columnas_categoricas_one_hot])
df_one_hot = pd.DataFrame(data_codificada_one_hot, columns=onehot.get_feature_names_out(), index=df_limpio.index)

# Codificación de frecuencia
encoder = CountEncoder(cols=columnas_categoricas_frecuencia)
df_frecuencia = encoder.fit_transform(df_limpio[columnas_categoricas_frecuencia])

# Escalado de columnas numéricas
scaler = RobustScaler()
df_numericas_escaladas = scaler.fit_transform(df_limpio[columnas_numericas])
df_numericas_escaladas = pd.DataFrame(df_numericas_escaladas, columns=columnas_numericas, index=df_limpio.index)

# Combinar todas las transformaciones en un solo DataFrame
df_transformado = pd.concat([df_limpio.drop(columnas_categoricas_one_hot + columnas_categoricas_frecuencia + columnas_numericas, axis=1),
                             df_one_hot, df_frecuencia, df_numericas_escaladas], axis=1)
```

```python
from sklearn.model_selection import train_test_split

# Asumiendo que 'X' son tus características y 'y_encoded' las etiquetas numéricas
X = df_transformado.drop(columns=['Survival_Range'])  # Definir características X
X_train, X_test, y_train, y_test = train_test_split(
    X,  # Usar las características X
    y_encoded,  # Usar las etiquetas numéricas
    test_size=0.2,  # Tamaño del conjunto de prueba
    random_state=42,  # Semilla para reproducibilidad
    stratify=y_encoded  # Estratificar por y_encoded para mantener la proporción de clases
)
```

Es importante recordar que, aunque SMOTE está diseñado para mejorar el trato de los modelos hacia las clases desbalanceadas, su objetivo principal no es necesariamente el de minimizar el sobreajuste en otras facetas.

Observamos que los modelos mantienen un nivel de rendimiento impresionantemente alto, tanto en los conjuntos de datos originales como en aquellos ajustados con SMOTE. Esto nos da a entender que ambos modelos muestran una fortaleza notable ante el desequilibrio de clases presente en este conjunto de datos particular.

Al implementar SMOTE, no hemos visto un incremento destacado en la precisión media de los modelos. En el caso de CatBoost, la precisión se mantiene casi inalterada, mientras que para XGBoost, se percibe un ligero descenso en la precisión media con el uso de SMOTE.

El hecho de que los puntajes de precisión rocen o alcancen el 100% en ambos tipos de conjuntos de datos podría sugerir que nos enfrentamos a un posible sobreajuste en los modelos. Aunque el propósito de aplicar SMOTE era optimizar el manejo de las clases menos representadas, la elevada precisión observada tanto en los datos originales como en los equilibrados nos indica que el desequilibrio de clases no era el único elemento influyente en la precisión excepcionalmente alta.
